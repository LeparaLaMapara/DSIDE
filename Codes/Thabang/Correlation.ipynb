# Is there a relationship between each measure on average?

```python
from pandas.plotting import scatter_matrix
dfmean = df_FinRatio.groupby(['Province']).mean()
scatter_matrix(dfmean[['waste_prop','repairs_PPE','currRatio','cash_bal','cash_costCov']],\
               alpha=1, figsize=(15, 10), diagonal='kde')
plt.show()
```

```python
start_date, end_date = 2013,2016
```

```python
for date in np.arange(start_date, end_date):
    df_w = df_FinRatio.loc[date].groupby("Province").mean()
    ax = df_w[['waste_prop']].plot(title='Wasted funds',figsize=(15,8))
    ax1 = df_w[['currRatio']].plot(title='Wasted funds',figsize=(15,8))
    ax2 = df_w[['cash_bal']].plot(title='Wasted funds',figsize=(15,8))
    ax3 = df_w[['cash_costCov']].plot(title='Wasted funds',figsize=(15,8))
    df_w[['waste_prop']].plot(title='Wasted funds',figsize=(15,8))
plt.xlabel('Year')
plt.ylabel('Waste ratio')
plt.show()
```
