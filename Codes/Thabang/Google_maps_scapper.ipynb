```python
def latlong(data):
    
    """
    Match the municipalitiy Name with lat/lon
    """
    
    APIKEY='token key here (use your own!!!)'
    gmaps = gm.Client(key=APIKEY)
    
    for loca in  data:
        loc = loca.split('/')[0]
        try:
            geocode_result = gmaps.geocode(loc)[0]
            latlon= geocode_result['geometry']['location']
        except IndexError: 
            latlon = {'lat':None,'lng' :None}

        data.loc[data.Name == loca,'Latitude'] = latlon['lat']
        data.loc[data.Name == loca,'Longitude'] = latlon['lng']
        
    return data    
```

```python

def Make_grid(xmin, xmax, ymin, ymax, gridx=200j, gridy=200j):
    """
    Create a regular 2D grid with N points in each dimension
    and modidy the x,y  to only have 100 points using ravel
    """
    xi, yi = np.mgrid[xmin:xmax:gridx, ymin:ymax:gridy]
    # points of  evauation
    positions = np.vstack([xi.ravel(), yi.ravel()])
    
    return  (positions , xi, yi)
```
