## Cash / Cost coverage 

This ration is used to indicate the municipalities ability to meet its operating commitments from cash and short-term investment without collecting any additional revenue.

\frac{cash - unspent conditional grants - overdraft + short-term investment}{Operational expenditure}

note:conditional grants, short-term investment still needs to be added

```python
#Cash receipts
cash_receipts = cash[cash["label"] == "Total Cash Receipts by Source"]
cash_receipts = cash_receipts[['Unnamed: 0', 'mun_code', 'total_amount']]
cash_receipts = cash_receipts.rename(columns = {'total_amount': 'cash_rec'})
cash_receipts.head()
```

## Fruitless and irregular expenditure - Efficient spending measure

This ratio has been formulated to compare the fruitless expenditure to the total expenditure.

```python
wastedExp = pd.read_csv("/Users/anrismith/Downloads/DSIDE-master 2/Municipal data/Monthly/fruitless_expenditure.csv")
wastedExp #wasted Exp only till 2015

#still yearly..
OpExp = OpExp[OpExp['date'] != 2016]
```

```python
fruitless = wastedExp[wastedExp['label'] == 'fruitless']
irregular = wastedExp[wastedExp['label'] == 'irregular']
unauthorised = wastedExp[wastedExp['label'] == 'unauthorised']
```

```python
waste = pd.DataFrame()
#waste['date'] = wastedExp['Unnamed: 0']
#waste['mun_code'] = wastedExp['mun_code']

waste = fruitless[['date', 'mun_code', 'total_amount']]
waste = waste.rename(columns = {'total_amount': 'fruitless'})
waste = waste.merge(irregular, on = ['date', 'mun_code'])
waste = waste.rename(columns = {'total_amount': 'irregular'})

waste = waste[['date', 'mun_code', 'fruitless', 'irregular' ]]

waste = waste.merge(unauthorised, on = ['date', 'mun_code'])
waste = waste.rename(columns = {'total_amount': 'unauthorised'})

waste = waste[['date', 'mun_code', 'fruitless', 'irregular', 'unauthorised']]

waste['total_waste'] = waste['fruitless'] + waste['irregular'] + waste['unauthorised']

waste = waste.merge(OpExp, on = ['date', 'mun_code'])

# THE BIGGER THIS MEASURE THE MORE EFFICIENT THE SPENDING
waste['Efficient_spending'] = 1 - (waste['fruitless'] + waste['irregular'] + waste['unauthorised']) / waste['opExp']
waste
```
