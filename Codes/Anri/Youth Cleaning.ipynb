# Youth Data Cleaning

```python
import pandas as pd

citizenship = pd.read_csv("/home/dside/Documents/GitHub/Youth data/YOUTH_CITIZENSHIP_South_Africa.csv")
prov_link = pd.read_csv("prov_link.csv")

citizenship
```

```python
def prov_gen(target, dictionary):

    prov = pd.DataFrame()
    target_codes = target.geo_code
    
    for i in range(0,len(target)): #for each row in the original data set
        for j in range(0,len(dictionary)):# find the correct province
            if target_codes.iloc[i] == dictionary.iloc[j,1]:
                prov = prov.set_value(len(prov),"province", dictionary.iloc[j,2])
    target["province"] = prov
    return target


citizenship = prov_gen(citizenship, prov_link)
citizenship.head()
```

```python
citizenship.to_csv("YOUTH_CITIZENSHIP_South_Africa.csv")                            
```

```python
unemployment_gender = pd.read_csv("YOUTH_EMPLOYMENT_STATUS_GENDER_South_Africa.csv")
unemployment_sector = pd.read_csv("YOUTH_EMPLOYMENT_SECTOR_GENDER_South_Africa.csv")
unemployment_training = pd.read_csv("YOUTH_EMPLOYMENT_EDUCATION_TRAINING_GENDER_South_Africa.csv")
```

```python
unemployment_gender = prov_gen(unemployment_gender, prov_link)
unemployment_sector = prov_gen(unemployment_sector, prov_link)
unemployment_training = prov_gen(unemployment_training, prov_link)
```
