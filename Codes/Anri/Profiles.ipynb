# Municipal and Youth profile 

```python
import pandas as pd
import statistics
```

```python
waste = pd.read_csv('/Users/anrismith/Downloads/DSIDE_HomeWork/MunicipalData/Wasteful_expenditure.csv')

IncExp = pd.read_csv("/Users/anrismith/Downloads/DSIDE_HomeWork/MunicipalData/income_expenditure.csv")
OpExp = IncExp[IncExp['label'] == 'Total Operating Expenditure']
OpExp = OpExp[OpExp['Unnamed: 0'] != 2016] # dont have 2016 data for waste
#OpExp = OpExp.rename(columns = {'Unnamed: 0': 'date'})
```

```python
fruitless = waste[waste['label'] == 'fruitless']
irregular = waste[waste['label'] == 'irregular']
unauthorised = waste[waste['label'] == 'unauthorised']
```

```python
fruitless
```

Notice that some of the wasted expenditure was NaN. These values were replaced by zeros

```python
Efficiency = pd.DataFrame()

Efficiency = fruitless[['Unnamed: 0', 'mun_code', 'total_amount']]
Efficiency = Efficiency.rename(columns = {'total_amount': 'fruitless'})
Efficiency = Efficiency.merge(irregular, on = ['Unnamed: 0', 'mun_code'])
Efficiency = Efficiency.rename(columns = {'total_amount': 'irregular'})
Efficiency= Efficiency[['Unnamed: 0', 'mun_code', 'fruitless', 'irregular']]

Efficiency = Efficiency.merge(unauthorised, on = ['Unnamed: 0', 'mun_code'])
Efficiency = Efficiency.rename(columns = {'total_amount': 'unauthorised'})

Efficiency = Efficiency[['Unnamed: 0', 'mun_code', 'fruitless', 'irregular', 'unauthorised']]
Efficiency['total_waste'] = Efficiency['fruitless'] + Efficiency['irregular'] + Efficiency['unauthorised']

Efficiency = Efficiency.merge(OpExp, on = ['Unnamed: 0', 'mun_code'])
Efficiency = Efficiency.rename(columns = {"total_amount":"OpExp"})
Efficiency = Efficiency[['Unnamed: 0', 'mun_code', 'fruitless', 'irregular', 'unauthorised', 'total_waste', 'OpExp']]

#Efficiency.fillna(0)


Efficiency
```

```python
# THE BIGGER THIS MEASURE THE MORE EFFICIENT THE SPENDING

Efficiency['Efficient_ratio'] = Efficiency['total_waste'] / Efficiency['OpExp']
Efficiency
```

```python
Eff_mean = statistics.mean(Efficiency['Efficient_ratio'])
Eff_sd = statistics.stdev(Efficiency['Efficient_ratio'])
Eff_max = max(Efficiency['Efficient_ratio'])
Eff_min = min(Efficiency['Efficient_ratio'])

#normalising the efficiency measure
Efficiency['Efficient_ratio'] = (Efficiency['Efficient_ratio'] - Eff_min) / (Eff_max - Eff_min)
 
Efficiency['Efficient_measure'] = 1 - Efficiency['Efficient_ratio'] 
del Efficiency['fruitless']
del Efficiency['irregular']
del Efficiency['unauthorised']
Efficiency
```

```python
Efficiency.shape # 829 rows
```

```python
# Wellbeing mesure
wellbeing = pd.read_csv('/Users/anrismith/Downloads/DSIDE_HomeWork/Youth data/Wellfare.csv')
del wellbeing['electricity_access']
del wellbeing['toilet_access']
del wellbeing['water_access']
del wellbeing['internet_access']
wellbeing
```

```python
Eff_Well = Efficiency.merge(wellbeing, how = 'left', on = ['mun_code'])
Eff_Well = Eff_Well.drop_duplicates()
Eff_Well.drop_duplicates()
del Eff_Well['total_waste']
del Eff_Well['OpExp']
del Eff_Well['Efficient_ratio']
Eff_Well
```

```python
Eff_Well.shape #829 rows
```

```python
# Calculating the profiles
#across time

profile=[]

for i in range(0,len(Eff_Well)):
    if Eff_Well['Efficient_measure'].iloc[i] >= 0.5 and Eff_Well['wellfare_measure'].iloc[i] >= 0.5:
        profile.append(1)
    if Eff_Well['Efficient_measure'].iloc[i] >= 0.5 and Eff_Well['wellfare_measure'].iloc[i] < 0.5:
         profile.append(2)
    if Eff_Well['Efficient_measure'].iloc[i] < 0.5 and Eff_Well['wellfare_measure'].iloc[i] < 0.5:
         profile.append(3)
    if Eff_Well['Efficient_measure'].iloc[i] < 0.5 and Eff_Well['wellfare_measure'].iloc[i] >= 0.5:
         profile.append(4)
            
```

```python
#profile
profile = pd.Series(profile)
#profile_data = pd.read_csv('/home/dside/Documents/GitHub/Analysis/PROFILE.csv')
Eff_Well['profile'] = profile
Eff_Well
```

```python
Eff_Well.to_csv('/Users/anrismith/Downloads/DSIDE_HomeWork/Youth data/PROFILE.csv')
```

```python
avg_Eff_Well = Eff_Well.groupby(['mun_code']).mean()
avg_Eff_Well
```

```python
#average profile

avg_profile = []

for i in range(0,len(avg_Eff_Well)):
    if avg_Eff_Well['Efficient_measure'].iloc[i] >= 0.5 and avg_Eff_Well['wellfare_measure'].iloc[i] >= 0.5:
       avg_profile.append(1)
    if avg_Eff_Well['Efficient_measure'].iloc[i] >= 0.5 and avg_Eff_Well['wellfare_measure'].iloc[i] < 0.5:
         avg_profile.append(2)
    if avg_Eff_Well['Efficient_measure'].iloc[i] < 0.5 and avg_Eff_Well['wellfare_measure'].iloc[i] < 0.5:
         avg_profile.append(3)
    if avg_Eff_Well['Efficient_measure'].iloc[i] < 0.5 and avg_Eff_Well['wellfare_measure'].iloc[i] >= 0.5:
         avg_profile.append(4)
```

```python
avg_profile = pd.Series(avg_profile)
#avg_profile
```

```python
avg_Eff_Well['avg_profile'] = avg_profile.values
del avg_Eff_Well['profile']
del avg_Eff_Well['Unnamed: 0']
avg_Eff_Well
```
