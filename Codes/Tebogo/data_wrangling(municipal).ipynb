```python
    
def latlon_match(data1,data2):
    
        """
        Match the municipalitiy Name with lat/lon
        """
        
        data1 = data1.reset_index()
        lat = []
        lon = []
        for i in np.arange(data1.shape[0]):
            for j in np.arange(data2.shape[0]):


                 if data1.loc[i,"mun_code"] == data2.loc[j,"Code"]:
                    lat.append(data2.loc[j,"Latitude"])
                    lon.append(data2.loc[j,"Longitude"])
                    

            
        dlat = pd.DataFrame(lat)
        dlat = pd.DataFrame(lon)
        data1['Latitude'] = dlat
        data1['Longitude'] = dlat
        data1 = data1.set_index('index')
        data1.index.name = None
       
        return data1
```
